```python
%matplotlib inline
import convokit
import matplotlib.pyplot as plt
```

```python
# load the corpus
corpus = convokit.Corpus(filename=convokit.download("reddit-corpus"))
```

```python
# create a hyperconvo object and use it to extract features
hc = convokit.HyperConvo(corpus)
threads_feats = hc.retrieve_feats()

# convenience method: embed threads into a low-dimensional space and group them into communities
pts, labels = hc.embed_communities(threads_feats, "subreddit", method="tsne")
```

```python
# plot
xs, ys = zip(*pts)
plt.scatter(xs, ys)
for i, txt in enumerate(labels):
    plt.annotate(txt, (xs[i], ys[i]))
plt.show()
```
