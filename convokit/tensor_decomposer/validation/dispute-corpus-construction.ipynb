```python
from convokit import Corpus, Utterance, Speaker
import json
```

```python
import os
os.listdir('dispute_wikipedia')
```

```python
with open('dispute_wikipedia/dispute_discussions.json', 'r') as f:
    disputes = json.load(f)
    
dispute_utts = []
for discussion in disputes:
    for utt_info in discussion['_lines']:
        speaker = Speaker(id=(utt_info['_user'] or '[MISSING]'))
        reply_to = None if utt_info['_line_number'] == 0 else f"{discussion['_discussion_ID']}-{utt_info['_line_number']-1}-disputed"
        utt = Utterance(id=f"{discussion['_discussion_ID']}-{utt_info['_line_number']}-disputed", 
                        speaker=speaker, text=utt_info['_text'],
                       timestamp=utt_info['_timestamp'], reply_to=reply_to, conversation_id=f"{discussion['_discussion_ID']}-0-disputed")
        dispute_utts.append(utt)
```

```python
with open('dispute_wikipedia/non_dispute_discussions.json', 'r') as f:
    nondisputes = json.load(f)
    
non_dispute_utts = []
for discussion in nondisputes:
    for utt_info in discussion['_lines']:
        speaker = Speaker(id=(utt_info['_user'] or '[MISSING]'))
        reply_to = None if utt_info['_line_number'] == 0 else f"{discussion['_discussion_ID']}-{utt_info['_line_number']-1}"
        utt = Utterance(id=f"{discussion['_discussion_ID']}-{utt_info['_line_number']}", 
                        speaker=speaker, text=utt_info['_text'],
                       timestamp=utt_info['_timestamp'], reply_to=reply_to, conversation_id=f"{discussion['_discussion_ID']}-0")
        non_dispute_utts.append(utt)
```

```python
convo_id_to_metadata = dict()
for discussion in disputes:
    convo_id = f"{discussion['_discussion_ID']}-0-disputed"
    metadata = dict()
    for k in ['_discussion_name', '_page_ID', '_page_name']:
        metadata[k[1:]] = discussion[k]
    metadata['has_dispute'] = True
    convo_id_to_metadata[convo_id] = metadata
    
for discussion in nondisputes:
    convo_id = f"{discussion['_discussion_ID']}-0"
    metadata = dict()
    for k in ['_discussion_name', '_page_ID', '_page_name']:
        metadata[k[1:]] = discussion[k]
    metadata['has_dispute'] = False
    convo_id_to_metadata[convo_id] = metadata
```

```python
for convo in corpus.iter_conversations():
    if not convo.check_integrity(verbose=False):
        convo.check_integrity(verbose=True)
```

```python
corpus = Corpus(utterances=dispute_utts + non_dispute_utts)
```

```python
corpus.dump(name='Wikipedia-Disputes-Corpus', base_path='.')
```

```python
os.listdir()
```

```python
for convo in corpus.iter_conversations():
    convo.meta.update(convo_id_to_metadata[convo.id])
```

```python
corpus.print_summary_stats()
```
